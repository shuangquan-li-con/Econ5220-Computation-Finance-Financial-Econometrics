```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
from scipy import stats
from google.colab import files
```

```python
df=pd.read_excel('hprice3.xlsx')
```

```python
df.columns
```

```python
df_1981=df[df['  1. year                     1978, 1981']==1981]
```

```python
y=df_1981['  8. price                    selling price']
```

```python
x=df_1981[['  2. age                      age of house', '  5. cbd                      dist. to central bus. dstrct, feet', '  6. inst                     dist. to interstate, feet',' 10. area                     square footage of house',' 11. land                     square footage lot','  9. rooms                    # rooms in house',' 12. baths                    # bathrooms']]
x.columns = ['age', 'cbd', 'inst', 'area', 'land', 'rooms', 'baths']
```

```python
x=sm.add_constant(x)
```

```python
model = sm.OLS(y,x).fit()
```

```python
print(model.summary())
```

```python
f_test = model.f_test([
    'land - area = 0',
    'rooms - baths = 0'
])
print(f_test)
```
