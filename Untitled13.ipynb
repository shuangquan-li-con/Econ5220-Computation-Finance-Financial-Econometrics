```python
df=pd.read_excel("data for HW1.xlsx")
```

```python
df.head()
```

```python
prices = df[['Date','IBM','SPY (ETF)']].set_index('Date')
```

```python
import numpy as np
returns = prices.pct_change()
returns = returns.replace([np.inf,-np.inf],np.nan)
returns = returns.dropna()
```

```python
y=returns['IBM']
x=sm.add_constant(returns['SPY (ETF)'])

```

```python
model=sm.OLS(y,x).fit()
```

```python
print(model.summary())
```

```python
beta_hat=model.params['SPY (ETF)']
se_beta=model.bse['SPY (ETF)']
```

```python
t_stat=(beta_hat-1)/se_beta
p_value=stats.t.cdf(t_stat,model.df_resid)
```
