```python
df=pd.read_excel('data for HW2.xlsx')
```

```python
for i in range(1,5):
  df[f'SPY_lag{i}']=df['SPY'].shift(i)
```

```python
df=df.replace([np.inf,-np.inf],np.nan)
```

```python
y=df['SPY']
x=df[['SPY_lag1','SPY_lag2','SPY_lag3','SPY_lag4']]
x=sm.add_constant(x)
```

```python
model_weak=sm.OLS(y,x, missing='drop').fit()
```

```python
print(model_weak.summary())
```

```python
f_test_weak=model_weak.f_test(
    "SPY_lag1=SPY_lag2=SPY_lag3=SPY_lag4=0"
)
print(f_test_weak)
```
